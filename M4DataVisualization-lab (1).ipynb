### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here

Query = "select *, count(*) as Count from DatabaseDesireNextYear \
Group by DatabaseDesireNextYear \
Order by Count desc limit 5"
df = pd.read_sql_query(Query, conn)

data = df['Count']
labels = df['DatabaseDesireNextYear']

plt.pie(data, labels = labels, autopct='%1.1f%%', explode = (0.1, 0, 0, 0, 0), shadow = True)
plt.show()
```

```python
# In the list of most popular languages respondents wish to learn next year, what is the rank of Python?

Query = "select *, count(*) as Count from LanguageDesireNextYear \
Group by LanguageDesireNextYear \
Order by Count desc limit 5"
df = pd.read_sql_query(Query, conn)

print(df)

data = df['Count']
labels = df['LanguageDesireNextYear']

plt.pie(data, labels = labels, autopct='%1.1f%%', explode = (0, 0, 0.1, 0, 0), shadow = True)
plt.show()
```

```python
# How many respondents indicated that they currently work with 'SQL'?

Query = "select *, count(*) as Count from LanguageWorkedWith \
Group by LanguageWorkedWith \
Order by Count desc limit 5"
df = pd.read_sql_query(Query, conn)

print(df)

data = df['Count']
labels = df['LanguageWorkedWith']

plt.pie(data, labels = labels, autopct='%1.1f%%', explode = (0, 0, 0.1, 0, 0), shadow = True)
plt.show()
```

```python
# How many respondents indicated that they work on 'MySQL' only?

Query = "select * from DatabaseWorkedWith \
Order by Respondent"
df = pd.read_sql_query(Query, conn)

counts = df.groupby('Respondent').count() #counts: number databases each respondent has worked with

uniques = []  # Initialize the list to store unique respondents

for i in range(len(counts)):
    if counts.values[i] == 1:
        uniques.append(counts.index[i]) #uniques: respondents who have worked with one database

uniq_mysql = 0

for j in range(len(df)):
    if df['Respondent'][j] in uniques:
        if df['DatabaseWorkedWith'][j] == 'MySQL':
            uniq_mysql = uniq_mysql + 1 #uniq_mysql: respondents who have worked with MySQL only

uniq_mysql
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here

Query = "select WorkWeekHrs, CodeRevHrs, Age  from master \
where Age between 30 and 35"
df = pd.read_sql_query(Query, conn)

df_median = df[['WorkWeekHrs', 'CodeRevHrs']].median()
df1 = pd.DataFrame(df_median).transpose()

df1.plot.bar(stacked=True)

plt.xlabel('Age 30-35')
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
Query = "select ConvertedComp, Age from master \
where Age between 45 and 60"
df = pd.read_sql_query(Query, conn)


medians = df.groupby('Age')['ConvertedComp'].median()
d1 = pd.DataFrame(medians)

sns.lineplot(data=d1, x='Age', y = 'ConvertedComp')

plt.show()
```

```python
# Based on the Line Chart of Age and median ConvertedComp for the age group 25-30,
# what kind of correlation do you see between Age and median ConvertedComp?

Query = "select ConvertedComp, Age from master \
where Age between 25 and 30"
df = pd.read_sql_query(Query, conn)


medians = df.groupby('Age')['ConvertedComp'].median()
d1 = pd.DataFrame(medians)

sns.lineplot(data=d1, x='Age', y = 'ConvertedComp')

plt.show()

# Positive correlation
```
